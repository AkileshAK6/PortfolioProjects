# BMI Calculator

```python
name= input("Enter your name: ")

weight= int(input("Enter you weight in pounds: "))

height= int(input("Enter you height in inches: "))

BMI = (weight * 703) / (height * height)

print(BMI)

if BMI>0:
    if(BMI<18.5):
        print(name+", you are Underweight, You need to eat more calorie dense foods and eat multiple times a day")
    elif(BMI<=24.9):
        print(name+", you are Normal Weight")
    elif(BMI<=29.9):
        print(name+", you are Overweight, You need to eat less and stop sitting and watching series")
    elif(BMI<=34.9):
        print(name+", you are Obese")
    elif(BMI<=39.9):
        print(name+", you are Severely Obese")
    else:
        print(name+", you are Morbidly Obese")   
```
